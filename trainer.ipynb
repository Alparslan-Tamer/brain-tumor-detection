```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

import glob
```

```python
model = YOLO("yolov8n.pt")
```

```python
model.train(data='/home/atamer/brain-tumor-detection/dataset/data.yaml', epochs=50, batch=16, imgsz=512)
```

```python
# import best weights
model = YOLO("/home/atamer/brain-tumor-detection/runs/detect/train8/weights/best.pt")
```

```python
val_image_paths = glob.glob("/home/atamer/brain-tumor-detection/dataset/validation/images/*.jpg")
val_label_paths = glob.glob("/home/atamer/brain-tumor-detection/dataset/validation/labels/*.txt")

val_image_paths.sort()
val_label_paths.sort()

random_index = np.random.randint(0, len(val_image_paths))

# load test image
results = model([val_image_paths[random_index]])

labels = []
# load label from text file
with open(val_label_paths[random_index], "r") as file:
    lines = file.readlines()

    for line in lines:
        line = line.strip().split(" ")
        label = int(line[0])
        labels.append(label)

    
print(val_label_paths[random_index])
print(val_image_paths[random_index])

for result in results:
    print("prediction: ", *result.boxes.cls.cpu().numpy())
    print("ground truth: ", labels)
    result.show()
```
